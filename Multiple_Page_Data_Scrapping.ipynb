```python
# required libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from random import randint
```

```python
url = 'https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser') # to get the content of the webpage in html format
```

```python
response
```

```python
#create empty list to store extracted data
movie_name = []
year = []
rating = []
time = []
metascore = []
votes = []
gross = []
description = []
director = []
stars = []
```

```python
#creating an array of values for dynamic webpage
pages = np.arange(1,1000,100)
```

```python
for page in pages:
    page = response
    soups = soup
    movie_data = soup.findAll('div', attrs = {'class': 'lister-item mode-advanced'})
    sleep(randint(2,8))
    
    for store in movie_data:
        name = store.h3.a.text
        movie_name.append(name)
        
        release_year = store.h3.find('span', class_ = "lister-item-year text-muted unbold").text
        year.append(release_year)
        
        runtime = store.p.find("span", class_ = 'runtime').text
        time.append(runtime)
        
        rate = store.find('div', class_ = "inline-block ratings-imdb-rating").text.replace('\n', '')
        rating.append(rate)
        
        score = store.find('span', class_ = "metascore").text if store.find('span', class_ = "metascore") else "^^^^"
        metascore.append(score)
        
        
        value = store.find_all('span', attrs = {'name': "nv"})
        
        vote = value[0].text
        votes.append(vote)
        
        gross_value = value[1].text if len(value)>1 else '$$$$'
        gross.append(gross_value)
         
        describe = store.find_all('p', class_ = 'text-muted')
        describe1 = describe[1].text.replace('\n', '') if len(describe) >1 else '####'
        description.append(describe1)
        
        cast = store.find("p", class_ = '')
        cast = cast.text.replace('\n', '').split('|')
        cast = [x.strip() for x in cast]
        cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
        director.append(cast[0])
        stars.append([x.strip() for x in cast[1].split(",")])
```

```python
# Create data frame
df_movie = pd.DataFrame({'Name of movie': movie_name, 'Release year': year, 'Runtime': time, 'Movie Rating': rating, 'Metascore': metascore, 'Votes': votes, 'Gross': gross, 'Description': description, 'Director': director, 'Star': stars})
```

```python
#save as csv file
df_movie.to_csv('Data Scrapping Multiple Pages on IMDb.csv', index=False)
```

```python
# save as Excel file
df_movie.to_excel('Data Scrapping Multiple Pages on IMDb.xlsx')
```
